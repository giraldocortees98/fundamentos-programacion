## Ejercicio 1 · Contador de palabras ignorando puntuación

Implementa `contar_palabras(texto)` que reciba un string con puntuación y devuelva un dict `{palabra: frecuencia}`. Debe ignorar la puntuación y convertir todo a minúsculas antes de contar.

Puntuación a eliminar: `.,;:!?"'()`

**Ejemplo:** `contar_palabras("Hola, mundo. Hola!")` → `{"hola": 2, "mundo": 1}`

Pista: itera sobre el texto y reemplaza cada carácter de puntuación por espacio,
luego divide por espacios y filtra strings vacíos.

```python
def contar_palabras(texto):
    puntuacion = set('.,;:!?"()')
    limpio = ''
    for c in texto.lower():
        if c in puntuacion:
            limpio += ' '
        else:
            limpio += c
    conteo = {}
    for p in limpio.split():
        if p:
            conteo[p] = conteo.get(p, 0) + 1
    return conteo
```

```python
# === Tests visibles · Ejercicio 1 ===
assert contar_palabras('Hola, mundo. Hola!') == {'hola': 2, 'mundo': 1}
assert contar_palabras('python') == {'python': 1}
print("✅ Ejercicio 1: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 1 ===
assert contar_palabras('') == {}
assert contar_palabras('a, a, a!') == {'a': 3}
assert contar_palabras('El sol sale.') == {'el': 1, 'sol': 1, 'sale': 1}
print("✅ Ejercicio 1: tests adicionales superados.")
```

## Ejercicio 2 · Índice invertido de posiciones

Implementa `indice_invertido(texto)` que reciba un string y devuelva un dict `{palabra: [lista_de_posiciones]}` donde cada posición es el índice (comenzando en 0) de la palabra en el texto.

**Ejemplo:**
`indice_invertido("el sol sale el sol")`
→ `{"el": [0, 3], "sol": [1, 4], "sale": [2]}`

El texto ya llega en minúsculas y sin puntuación.

```python
def indice_invertido(texto):
    indice = {}
    for pos, palabra in enumerate(texto.split()):
        if palabra not in indice:
            indice[palabra] = []
        indice[palabra].append(pos)
    return indice
```

```python
# === Tests visibles · Ejercicio 2 ===
assert indice_invertido('el sol sale el sol') == {'el': [0, 3], 'sol': [1, 4], 'sale': [2]}
print("✅ Ejercicio 2: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 2 ===
assert indice_invertido('') == {}
assert indice_invertido('a') == {'a': [0]}
assert indice_invertido('x y x y x') == {'x': [0, 2, 4], 'y': [1, 3]}
print("✅ Ejercicio 2: tests adicionales superados.")
```

## Ejercicio 3 · Primer carácter no repetido

Implementa `primer_no_repetido(cadena)` que devuelva el **primer carácter** de la cadena que aparece exactamente una vez. Si no existe, devuelve `None`.

**Ejemplo:** `primer_no_repetido("aabbcde")` → `'c'`

**Ejemplo:** `primer_no_repetido("aabb")` → `None`

Pista: construye primero un dict de frecuencias, luego recorre la cadena
buscando el primer carácter con frecuencia 1.

```python
def primer_no_repetido(cadena):
    freq = {}
    for c in cadena:
        freq[c] = freq.get(c, 0) + 1
    for c in cadena:
        if freq[c] == 1:
            return c
    return None
```

```python
# === Tests visibles · Ejercicio 3 ===
assert primer_no_repetido('aabbcde') == 'c'
assert primer_no_repetido('aabb') is None
print("✅ Ejercicio 3: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 3 ===
assert primer_no_repetido('a') == 'a'
assert primer_no_repetido('') is None
assert primer_no_repetido('abacabad') == 'c'
assert primer_no_repetido('aabbcc') is None
print("✅ Ejercicio 3: tests adicionales superados.")
```

## Ejercicio 4 · Histograma de valores

Implementa `histograma(valores)` que reciba una lista de números enteros y devuelva un dict `{valor: frecuencia}` (igual a un contador de frecuencias).

Luego imprime el histograma con barras de asteriscos, ordenado por valor.

La función solo devuelve el dict; la impresión es parte del análisis.

**Ejemplo:** `histograma([1, 2, 1, 3, 2, 1])` → `{1: 3, 2: 2, 3: 1}`

```python
def histograma(valores):
    h = {}
    for v in valores:
        h[v] = h.get(v, 0) + 1
    return h
```

```python
# === Tests visibles · Ejercicio 4 ===
assert histograma([1, 2, 1, 3, 2, 1]) == {1: 3, 2: 2, 3: 1}
assert histograma([]) == {}
print("✅ Ejercicio 4: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 4 ===
assert histograma([5]) == {5: 1}
assert histograma([1, 1, 1]) == {1: 3}
assert histograma([3, 1, 2, 1, 3, 3]) == {3: 3, 1: 2, 2: 1}
print("✅ Ejercicio 4: tests adicionales superados.")
```

## Ejercicio 5 · Total de compras con manejo de clave inexistente

Implementa `total_compras(precios, lista_compras)` que reciba:
- `precios`: dict `{producto: precio}`
- `lista_compras`: lista de nombres de producto

Devuelve la suma de los precios de los productos en `lista_compras`.
Si un producto **no está en el catálogo**, su precio se trata como 0
(no lanza error).

**Ejemplo:**
`total_compras({'laptop': 2500, 'teclado': 120}, ['laptop', 'mouse'])`
→ `2500`  (mouse no está, su precio es 0)

```python
def total_compras(precios, lista_compras):
    total = 0
    for producto in lista_compras:
        total += precios.get(producto, 0)
    return total
```

```python
# === Tests visibles · Ejercicio 5 ===
assert total_compras({'laptop': 2500, 'teclado': 120}, ['laptop', 'mouse']) == 2500
assert total_compras({'a': 10, 'b': 20}, ['a', 'b']) == 30
print("✅ Ejercicio 5: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 5 ===
assert total_compras({}, ['x', 'y']) == 0
assert total_compras({'x': 5}, []) == 0
assert total_compras({'a': 100}, ['a', 'a', 'a']) == 300
print("✅ Ejercicio 5: tests adicionales superados.")
```

## Ejercicio 6 · Comparar dos conjuntos

Implementa `comparar_conjuntos(a, b)` que reciba dos listas (o sets) y devuelva una **tupla** `(solo_en_a, solo_en_b, en_ambos)` donde cada componente es un **set** ordenado.

**Ejemplo:**
`comparar_conjuntos([1, 2, 3, 4], [3, 4, 5, 6])`
→ `({1, 2}, {5, 6}, {3, 4})`

```python
def comparar_conjuntos(a, b):
    sa = set(a)
    sb = set(b)
    return (sa - sb, sb - sa, sa & sb)
```

```python
# === Tests visibles · Ejercicio 6 ===
assert comparar_conjuntos([1, 2, 3, 4], [3, 4, 5, 6]) == ({1, 2}, {5, 6}, {3, 4})
print("✅ Ejercicio 6: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 6 ===
assert comparar_conjuntos([], []) == (set(), set(), set())
assert comparar_conjuntos([1, 2], []) == ({1, 2}, set(), set())
assert comparar_conjuntos([1], [1]) == (set(), set(), {1})
assert comparar_conjuntos([1, 1, 2], [2, 3]) == ({1}, {3}, {2})
print("✅ Ejercicio 6: tests adicionales superados.")
```

## Ejercicio 7 · Aplanar un dict anidado

Implementa `aplanar_dict(d)` que reciba un dict donde los valores son también dicts, y devuelva un dict plano con **tuplas** como claves.

**Ejemplo:**
`aplanar_dict({'a': {'x': 1, 'y': 2}, 'b': {'x': 3}})`
→ `{('a', 'x'): 1, ('a', 'y'): 2, ('b', 'x'): 3}`

Solo se aplana un nivel (el valor interno puede ser cualquier tipo).

```python
def aplanar_dict(d):
    resultado = {}
    for clave_ext, sub_dict in d.items():
        for clave_int, valor in sub_dict.items():
            resultado[(clave_ext, clave_int)] = valor
    return resultado
```

```python
# === Tests visibles · Ejercicio 7 ===
assert aplanar_dict({'a': {'x': 1, 'y': 2}, 'b': {'x': 3}}) == {('a', 'x'): 1, ('a', 'y'): 2, ('b', 'x'): 3}
print("✅ Ejercicio 7: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 7 ===
assert aplanar_dict({}) == {}
assert aplanar_dict({'k': {'v': 99}}) == {('k', 'v'): 99}
assert aplanar_dict({'x': {'a': 1}, 'y': {'b': 2}}) == {('x', 'a'): 1, ('y', 'b'): 2}
print("✅ Ejercicio 7: tests adicionales superados.")
```

## Ejercicio 8 · Moda con dict (O(n) vs O(n²) de la Clase 1)

Implementa `moda(lista)` que devuelva el elemento que **más veces** aparece en la lista. En caso de empate, devuelve el que aparece **primero** al recorrer la lista (no necesariamente el de menor valor).

**Esta vez usa un dict** para hacerlo en O(n), no en O(n²).

**Ejemplo:** `moda([1, 2, 2, 3, 3, 3])` → `3`

**Ejemplo:** `moda([4, 4, 5, 5])` → `4` (empate: gana el primero)

```python
def moda(lista):
    freq = {}
    orden = []
    for x in lista:
        if x not in freq:
            orden.append(x)
        freq[x] = freq.get(x, 0) + 1
    mejor = orden[0]
    for x in orden:
        if freq[x] > freq[mejor]:
            mejor = x
    return mejor
```

```python
# === Tests visibles · Ejercicio 8 ===
assert moda([1, 2, 2, 3, 3, 3]) == 3
assert moda([4, 4, 5, 5]) == 4
print("✅ Ejercicio 8: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 8 ===
assert moda([7]) == 7
assert moda(['a', 'b', 'a']) == 'a'
assert moda([1, 2, 3, 2, 1, 2]) == 2
print("✅ Ejercicio 8: tests adicionales superados.")
```
